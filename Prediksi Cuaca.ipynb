Deskripsi atribut digunakan untuk mempertimbangkan atribut mana yang akan di drop.
- lat: Garis lintang (latitude) lokasi cuaca.
- lon: Garis bujur (longitude) lokasi cuaca.
- timezone: Zona waktu lokasi cuaca.
- timezone_offset: Perbedaan waktu dari UTC (Coordinated Universal Time) dalam detik.
- dt: Timestamp (waktu) dari data cuaca, biasanya dalam format UNIX Epoch.
- sunrise: Waktu matahari terbit.
- sunset: Waktu matahari terbenam.
- moonrise: Waktu bulan terbit.
- moonset: Waktu bulan terbenam.
- moon_phase: Fase bulan pada waktu tertentu (misalnya, bulan purnama, bulan sabit, dll.).
- pressure: Tekanan udara, sering kali dalam hektopaskal (hPa).
- humidity: Kelembaban relatif udara dalam persentase.
- dew_point: Titik embun, yaitu suhu di mana udara jenuh dengan uap air.
- wind_speed: Kecepatan angin, sering kali dalam meter per detik.
- wind_deg: Arah angin dalam derajat.
- wind_gust: Kecepatan rata-rata angin dalam waktu tertentu.
- clouds: Persentase tutupan awan.
- pop: Probabilitas hujan (Probability of Precipitation).
- rain: Jumlah hujan yang diharapkan, sering kali dalam satuan milimeter (mm).
- uvi: Indeks sinar ultraviolet (UV Index).
- temp_day: Suhu maksimum pada siang hari.
- temp_min: Suhu minimum.
- temp_max: Suhu maksimum.
- temp_night: Suhu pada malam hari.
- temp_eve: Suhu pada petang hari.
- temp_morn: Suhu pada pagi hari.
- feels_like_day: Perasaan suhu maksimum.
- feels_like_night: Perasaan suhu malam.
- feels_like_eve: Perasaan suhu petang.
- feels_like_morn: Perasaan suhu pagi.
- weather_id: ID kode cuaca.
- weather_main: Kondisi cuaca utama (misalnya, Clear, Rain, Clouds).
- weather_description: Deskripsi cuaca yang lebih spesifik.
- weather_icon: Icon cuaca yang terkait.
- current: Status cuaca saat ini.
- location: Lokasi atau nama lokasi terkait data cuaca ini.

#### ZSCORE

```python
from scipy import stats 

sg_num = sg[['dt', 'sunrise', 'sunset','moonrise', 'moonset', 'moon_phase', 'pressure',
             'wind_speed', 'wind_gust', 'rain', 'uvi',
             'temp_day', 'temp_max', 'temp_night', 'temp_eve', 'feels_like_day',
             'feels_like_eve', 'current', 'weather_id']]
sg_num
```

```python
from scipy import stats 

stats.zscore(sg_num, nan_policy='omit')
```

```python
sg_zscore=sg_num[(np.abs(stats.zscore(sg_num, nan_policy='omit')) < 5).all(axis=1)]
sg_zscore.shape
```

```python
sg_zscore.plot(kind='box', figsize=(20,16))
plt.show()
```

```python
sg_zscore.nunique()
```

```python
unique_weather_id = sg_zscore['weather_id'].unique()

for weather_id in unique_weather_id:
    print("Weather ID:", weather_id)
```

### MODELLING

```python
# Deklarasi x dan y
X = sg_zscore.drop(['weather_id'], axis = 1)
y = sg_zscore['weather_id']
```

```python
# Split data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.27, random_state = 0)
X_train.shape, X_test.shape
```

```python
# Feature engineering
X_train.dtypes
```

```python
numerical = [col for col in X_train.columns if X_train[col].dtypes != 'O']
numerical
```

```python
X_train[numerical].isnull().sum()
```

```python
X_test[numerical].isnull().sum()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.naive_bayes import CategoricalNB
```

##### Logistic Regression

```python
# Logistik Regression
LRclassifier = make_pipeline(StandardScaler(), LogisticRegression())
LRclassifier.fit(X_train, y_train)
```

```python
y_pred = LRclassifier.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import accuracy_score
LRAcc = accuracy_score(y_pred,y_test)
print('LR accuracy: {:.2f}%'.format(LRAcc*100))
```

```python
# K-Fold Validation  (K=10)
from sklearn.model_selection import cross_val_score

scores = cross_val_score(LRclassifier, X_train, y_train, cv = 10, scoring='accuracy')

print('Cross-validation scores:{}'.format(scores))

print('MIN cross-validation score: {:.4f}'.format(scores.min()))
print('MAX cross-validation score: {:.4f}'.format(scores.max()))
print('AVG cross-validation score: {:.4f}'.format(scores.mean()))
```

```python
# Error Rate
from sklearn.metrics import accuracy_score
error_rate = 1 - accuracy_score(y_test, y_pred)
print('Error rate:', error_rate)
```

##### Random Forest

```python
# Random Forest
RFclassifier = make_pipeline(StandardScaler(), RandomForestClassifier(n_estimators = 100))
RFclassifier.fit(X_train, y_train)
```

```python
y_pred = RFclassifier.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import accuracy_score
RFAcc = accuracy_score(y_pred,y_test)
print('RF accuracy: {:.2f}%'.format(RFAcc*100))
```

```python
# K-Fold Validation  (K=10)
from sklearn.model_selection import cross_val_score

scores = cross_val_score(RFclassifier, X_train, y_train, cv = 10, scoring='accuracy')

print('Cross-validation scores:{}'.format(scores))

print('MIN cross-validation score: {:.4f}'.format(scores.min()))
print('MAX cross-validation score: {:.4f}'.format(scores.max()))
print('AVG cross-validation score: {:.4f}'.format(scores.mean()))
```
